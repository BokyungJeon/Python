```python
# 정규화 normalization

# Python 수치해석 라이브러리
import numpy as np

# 회귀 분석용 테스트 데이터 만들어주는 라이브러리
from sklearn.datasets import make_regression

X, y = make_regression(n_samples = 500, n_features = 2, n_informative = 2, noise = 5, random_state = 0)

print(X.shape)
print(y.shape)
```

```python
print("output range = [%f %f]" %(min(y), max(y)))
```

```python
# pip install keras
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
def model_build():
    model = Sequential()
    
    model.add(Dense(16, input_dim = 2, activation = "relu"))
    model.add(Dense(8, activation = "relu"))
    model.add(Dense(1))
    
    model.compile(loss='mean_squared_error', optimizer='adam')
    
    return model
```

```python
from keras.wrappers.scikit_learn import KerasRegressor
```

```python
YourModel = KerasRegressor(build_fn = model_build, epochs = 400, batch_size = 10, verbose = 1)
```

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits = 5)
```

```python
from sklearn.model_selection import cross_val_score

results = cross_val_score(YourModel, X, y, cv=kf)
```

```python
print("Final Cross Validation Loss:", abs(results.mean()))
```

```python
# l2는 오차에 대한 에러율을 낮추는데 활용한다.
# 오차의 제곱을 가지고 연산하므로 잘못된 정보를 판단할 가능성 자체가 적음
# 특히 음수와 양수 형태로 구성된 오차간의 정보 오판율은 굉장히 낮다.
# l2의 단점은 제곱이기때문에 분산값이 굉장히 커질 수 있다.
# 분산이 커지는 것이 싫다면 l1_l2를 사용하도록 한다.

# Dropout은 잡음noise에 대한 정보를 학습하는 과적합을 해결하기 위함
# 일정 수치 이하의 정보에 대해서는 걸러낸다.
# 이 수치를 너무 크게 줄 경우 실제 정보도 같이 증발되므로 적정한 수치값을 찾아내는 것이 매우 중요하다.
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
X, y = make_classification(n_samples = 500, n_features = 16, n_classes = 2)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
print(X.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
model = Sequential()
model.add(Dense(16, input_dim = 16, activation = 'relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(4, activation = 'tanh'))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(optimizer = 'sgd', loss = 'binary_crossentropy')

trainRes = model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 300, batch_size = 5, verbose = 0)

print("Test Loss =", model.evaluate(X_test, y_test))
```

```python
plt.plot(trainRes.history['loss'])
plt.plot(trainRes.history['val_loss'])
plt.ylim(0, 1)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train loss', 'test loss'], loc='upper right')
```

```python
# 훈련(학습)을 많이 했더니 훈련 로스는 잘나왔으나(거의 없다) 테스트 로스는 증가함. 
# 학습을 많이했더니 노이즈까지 학습했다는 뜻. 대표적인 과적합.(과적ㅎ바이 되어서 노이즈 정보까지 처리하는 상화)

# 해결책
# 1. 훈련 수를 줄이면 된다.(Early Stopping. 조기 중단)
# 2. l2를 사용해 오차의 제곱의 합을 가지고 연산
#    오차가 양수(+), 음수(-)로 인해 잘못된 판정을 할 수 있다.(합이 0)
#    이를 방지하기 위해 양쪽 모두 양수가 될 수 있는 제곱 형태를 취함
# 3. Dropout로 일정 확률 이하의 정보를 폐기처분
```

```python
from keras.callbacks import EarlyStopping
```

```python
model = Sequential()
model.add(Dense(16, input_dim = 16, activation = 'relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(4, activation = 'tanh'))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(optimizer = 'sgd', loss = 'binary_crossentropy')

# EarlyStopping을 사용한다면 Fit을 하기 이전에 어디서 중단할 것인지 정해줘야 한다.
# 콜백이란? 이벤트. 조건이 만족되면 콜백에 의해 fit이 중단되도록 구성.
#             monitor = val_loss: 테스트 로스를 감독함 / mode = 'min': 최소값을 감지함
es_callbak = EarlyStopping(monitor = 'val_loss', mode = 'min')

trainRes = model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 300, batch_size = 5, 
                     callbacks = [es_callbak], verbose = 0)

print("Test Loss =", model.evaluate(X_test, y_test))
```

```python
plt.plot(trainRes.history['loss'])
plt.plot(trainRes.history['val_loss'])
plt.ylim(0, 1)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train loss', 'test loss'], loc='upper right')
```
