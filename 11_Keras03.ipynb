```python
import matplotlib.patches as mpatches

from sklearn.datasets import make_blobs, make_circles
```

```python
def plot_data(pl, x, y):
    # 아래 make_circles로 만든 x,y에 대하여
    # y(참,거짓) == 0인 것들 중 x의(2차원정보)의 x좌표(,0) / y == 0인 애들 중에 x의 y좌표 (,1)
    pl.plot(x[y == 0, 0], x[y == 0, 1], 'ob', alpha = 0.5)
    pl.plot(x[y == 1, 0], x[y == 1, 1], 'xr', alpha = 0.5)
    pl.legend(['0', '1'])
    return pl
```

```python
# make_circles로 만든 값 x는 좌표(2차원정보), y는 노이즈 0.1에서 참과 거짓
x, y = make_circles(n_samples = 400, factor = 0.6, noise = 0.1, random_state = 42)

pl = plot_data(plt, x, y)
pl.show()

print(y)
```

```python
def plot_decision_bound(model, x, y):
    # x shape ->(400,2) 400개의데이터정보가2개씩
    # a = x, b = y 좌표로 최대값 최소값 설정
    amin, bmin = x.min(axis = 0) - 0.1
    amax, bmax = x.max(axis = 0) + 0.1
    print("amin = ", amin)
    print("bmax = ", amax)
    print("amin = ", amin)
    print("bmax = ", amax)
    
    # np.linspace()는 첫번째 인자에서 두번째 인자 사이의 값을 세번째 인자의 개수만큼 균일하게 쪼갠다.
    # 실제 데이터를 전부 포함할 수 있는 컴팩트한 사각형의 범위를 잡는다.
    hticks = np.linspace(amin, amax, 101)
    vticks = np.linspace(bmin, bmax, 101)
    
    # np.meshgrid() x축 정보(벡터), y축 정보(벡터)를 입력받아 사각형 영역에 반환한다.
    # 좌표 백터로부터 좌표 행렬을 반환한다.
    aa, bb = np.meshgrid(hticks, vticks)
    ab = np.c_[aa.ravel(), bb.ravel()]
    
    # model.predict(ab)를 사용하여 아래쪽에서 cntourf를 사용할 준비를 한다.
    # c라는 정보는 contour에서 활용하는 등고선의 기준 높이로 활용된다.
    c = model.predict(ab)
    z = c.reshape(aa.shape)
    
    # contourf(*arg, data=None, **kwargs) 등고선이 같은 애들의 범주를 잡아줌
    # cmp와 alpha는 **kwargs이 처리하고 aa, bb는 *arg이 처리
    # cmap == colormap을 bwr을 택할 경우 아래파랑 중간흰색 위에빨강
    # alpha값은 투명도를 의미. 낮을수록 투명도가 높다. 높을수록 색상이 진해진다줌
    plt.contourf(aa, bb, z, cmap = 'bwr', alpha = 0.2)
    # plot_data를 통해 실제 데이터를 그림
    plot_data(plt, x, y)
    return plt
```

```python
import pandas as pd

data = pd.read_csv("./Applied-Deep-Learning-with-Keras/Lesson02/bank_data_feats.csv", index_col = 0)
y = pd.read_csv("./Applied-Deep-Learning-with-Keras/Lesson02/bank_data_target.csv", index_col = 0)

x = data
```

```python
print("Num of Dataset =", data.shape[0])
print("Num of Feattures =", data.shape[1])
print(y.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense

def build_model():
    model = Sequential()
    model.add(Dense(16, input_dim = x.shape[1], activation = 'relu'))
    model.add(Dense(8, activation = 'relu'))
    model.add(Dense(1, activation = 'sigmoid'))
    
    model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
    
    return model
```

```python
import numpy as np
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

```python
seed = 1
np.random.seed(seed)
```

```python
n_folds = 5
epochs = 300
batch_size = 5

# 분류 작업을 할 때 사용하는 분류기 생성자
# 실질적으로 분류를 수행하는데 사용할 모델을 만드는 함수/ 몇회 반복/ 텐서큐브행렬(너비높이배치사이즈)나중에 설명
classifier = KerasClassifier(build_fn = build_model, epochs = epochs, batch_size = batch_size, verbose = 1)
```

```python
# 우리가 가지고 있는 Train, Test 데이터에 대한 검증 생성자
# Train, Test 총합 1000개 있다고 가정했을 때 
# 1. Train 0~799, Test 800~999
# 2. Train 200~999, Test 0~999
# 3. Train 0~199, 400~999, Test 200~399
# .....를 테스트해서 돌려주겠다

# 전체 구간을 5번으로 분할해서 검증하도록 만든다.
kfold = StratifiedKFold(n_splits = n_folds, shuffle = True, random_state = seed)
```

```python
# 위에서 만든 분류기와 검증기를 전달해서 x, y값에 대한 교차검증을 수행한다.
results = cross_val_score(classifier, x, y, cv = kfold)
```

```python
for f in range(n_folds):
    print("Accuracy at fold", f + 1, "=", results[f])
    
print("")
print("Final Cross Valid Test Accuracy:", results.mean())
print("Standard Deviation of Final Test Accuracy:", results.std())
```

```python
# default parameter는 만약 인자가 없으면 자동으로 해당인자를(parameter)를 지정된 녀석으로 주는 것
# 인자가 있으면 대체하게됨
def model_build1(activation = 'relu', optimizer = 'adam'):
    model = Sequential()
    
    model.add(Dense(8, input_dim = x.shape[1], activation = activation))
    model.add(Dense(8, activation = activation))
    model.add(Dense(8, activation = activation))
    model.add(Dense(1, activation = 'sigmoid'))
    
    model.compile(loss = 'binary_crossentropy', optimizer = optimizer, metrics = ['accuracy'])
    
    return model
```

```python
def model_build2(activation = 'relu', optimizer = 'adam'):
    model = Sequential()
    
    model.add(Dense(16, input_dim = x.shape[1], activation = activation))
    model.add(Dense(8, activation = activation))
    model.add(Dense(4, activation = activation))
    model.add(Dense(1, activation = 'sigmoid'))
    
    model.compile(loss = 'binary_crossentropy', optimizer = optimizer, metrics = ['accuracy'])
    
    return model
```

```python
def model_build3(activation = 'relu', optimizer = 'adam'):
    model = Sequential()
    
    model.add(Dense(8, input_dim = x.shape[1], activation = activation))
    model.add(Dense(4, activation = activation))
    model.add(Dense(1, activation = 'sigmoid'))
    
    model.compile(loss = 'binary_crossentropy', optimizer = optimizer, metrics = ['accuracy'])
    
    return model
```

```python
# seed는 이미되어있으니 필요없고
batch_size = 5
epochs = 200

results = []

models = [model_build1, model_build2, model_build3]
```

```python
# 인공신경망 모델에 대한 교차 검증하여 어떤 모델이 가장 성능이 좋은지를 판정한다.
for idx in range(len(models)):
    classifier = KerasClassifier(build_fn = models[idx], epochs = epochs, batch_size = batch_size, verbose = 1)
    
    kfold = StratifiedKFold( n_splits = n_folds, shuffle = True, random_state = seed)
    
    
    result = cross_val_score(classifier, x, y, cv = kfold)
    results.append(result)
```

```python
for idx in range(len(models)):
    print("Model", idx + 1, "Test Accuracy =", results[idx].mean())
```

```python
# Hyper Parameter 초매개변수
# 교차검증을 통해서 어떤 녀석을 활용하는 것이 가장 좋은지 판정할 수 있다.
optimizers = ['rmsprop', 'adam', 'sgd']
activations = ['relu', 'tanh']

for opt in range(len(optimizers)):
    for active in range(len(activations)):
        optimizer = optimizers[opt]
        activation = activations[active]
        
        classifier = KerasClassifier(build_fn = model_build1, epochs = 300, batch_size = 5, verbose = 1)
        
        kfold = StratifiedKFold(n_splits = n_folds, shuffle = True, random_state = seed)
        
        result = cross_val_score(classifier, x, y, cv = kfold)
        
        results.append(result)
```
